```python
import pandas as pd 
import numpy as np
from pathlib import Path
import plotly.express as px
```

```python
data_folder_path = Path().resolve().parent / 'DATA'

dframes ={}
# Iterate through all CSV files in the folder
for csv in data_folder_path.glob('*.csv'):
    df = pd.read_csv(csv)
    key = csv.stem
    dframes[key] = df
    
dframes.keys()
```

```python
dframes['movies']['genres'] = dframes['movies']['genres'].apply(
    lambda x: x.split('|')
)


userid_desc = dframes['ratings'].groupby('userId').agg(
    {
        'movieId':['nunique','unique'],
        'rating':['mean','std']
    }
)
userid_desc
```

```python
def get_user_rated_movies(user_id: int):
    """Return titles and genres for all movies rated by the given user."""
    ratings = dframes['ratings']
    movies = dframes['movies']
    # filter ratings and merge with movie metadata
    user_ratings = ratings[ratings['userId'] == user_id]
    merged = user_ratings.merge(
        movies[['movieId','title','genres']],
        on='movieId',
        how='left'
    )
    return merged[['movieId','title','genres','rating']]
```
